```python
from datasets import load_dataset
import pandas as pd
import random

# Load dataset
dataset = load_dataset("openai_humaneval")
num_samples = 164

# Create original dataframe
data = []
for i in range(num_samples):
    sample_data = {
        "task_id": dataset["test"][i]["task_id"],
        "prompt": dataset["test"][i]["prompt"],
        "canonical_solution": dataset["test"][i]["canonical_solution"],
        "test": dataset["test"][i]["test"],
        "entry_point": dataset["test"][i]["entry_point"]
    }
    data.append(sample_data)
original_df = pd.DataFrame(data)

# Mask words while preserving newlines
def mask_continuous_words_with_newlines(code, mask_ratio=0.1):
    lines = code.splitlines()
    masked_lines = []
    for line in lines:
        words = line.split()
        total_words = len(words)
        if total_words > 0:
            num_to_mask = max(1, int(round(total_words * mask_ratio)))
            start_index = random.randint(0, total_words - num_to_mask)
            for i in range(start_index, start_index + num_to_mask):
                words[i] = ''
        masked_lines.append(' '.join(words))
    return '\n'.join(masked_lines)

masked_prompts = [
    mask_continuous_words_with_newlines(code) for code in original_df['canonical_solution']
]
masked_df = pd.DataFrame(masked_prompts, columns=['masked_code'])

# Combine dataframes and save
final_df = pd.concat([original_df, masked_df], axis=1)
final_df.to_csv('masked_code_with_newlines.csv', index=False)

print("Processing complete. Saved to 'masked_code_with_newlines.csv'")
```
